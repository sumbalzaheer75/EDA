# Exploratory Data Analysis

## Size

```python
ax =sns.countplot(x='Size', data=df, palette='viridis')
```

```python
ax=sns.countplot(x='Size',data=df, palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
quantity = df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
sns.barplot(x='Size',y='Quantity',data=quantity, palette = 'pastel')
```

#### From above graph we can see most people buy M-Size

```python
# we're making histogram om Size
df['Size'].hist()
```

```python
# we convert Category to string
# we made a variable called column_data
# "bins" is used for distance between bars
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

#### From above graph you can see that most of the buyers are T-shirt

```python
# checking B2B Data using pie chart
B2B_Check = df['B2B'].value_counts()

# plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
plt.show()
```

#### From the above chart we can see that, maximum i.e 99.2% buyers are retailers and 0.8% are B2B buyers.

```python
# we use scatter plot to find how many data and how much and different types of size

x_data = df['Category']
y_data = df['Size']

# plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
# Plot count of cities by state

plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state', palette='Set2')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 states
top_10_states = df['ship-state'].value_counts().head(10)

# Plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x= 'ship-state', palette = 'Set2')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

#### From above graph you can see that most of the buyers are from Maharashtra state.
